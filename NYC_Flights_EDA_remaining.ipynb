```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data into a df
flights_df= pd.read_csv('C:\\Users\\ashwani\\Project_1_NYC-Flight data/flight_data.csv')
```

```python
#Fitering dataset for flights having arr_delay<=0
flights_df_filtered = flights_df[(flights_df.arr_delay<=0) & np.isfinite(flights_df['arr_time'])]  

#Grouping by carrier and getting the count
flights_df_filtered_grouped = flights_df_filtered.groupby([flights_df_filtered.carrier]).size().reset_index(name='OnTimeFlights')
print (flights_df_filtered_grouped)

#Plotting OnTime Flights for each Carrier
ind = np.arange(len(flights_df_filtered_grouped.carrier))
plt.bar(ind, flights_df_filtered_grouped.OnTimeFlights, color="green")
plt.ylabel('No of OnTimeFlights')
plt.xlabel('Carriers')
plt.title('OnTime Performance of Carriers')
plt.xticks(ind, flights_df_filtered_grouped.carrier)
plt.show()
```

```python
grouped_carrier_total = flights_df.groupby([flights_df.carrier]).size().reset_index(name='NoOfFlights')
print (grouped_carrier_total)

#Plotting Total No. of Flights for each Carrier
ind = np.arange(len(grouped_carrier_total.carrier))
plt.bar(ind, grouped_carrier_total.NoOfFlights, color="blue")
plt.ylabel('No of Flights')
plt.xlabel('Carriers')
plt.title('Total Flights per Carrier')
plt.xticks(ind, grouped_carrier_total.carrier)
plt.show()
```

```python
#Fitering dataset for flights having dep_delay>0
flights_df_filtered = flights_df[(flights_df.dep_delay>0) & np.isfinite(flights_df['dep_time'])]  

#Grouping by carrier and getting the count
flights_df_filtered_grouped = flights_df_filtered.groupby([flights_df_filtered.carrier]).size().reset_index(name='DepDelayedFLights')
print (flights_df_filtered_grouped)

#Plotting Departure Delayed Flights for each Carrier
ind = np.arange(len(flights_df_filtered_grouped.carrier))
plt.bar(ind, flights_df_filtered_grouped.DepDelayedFLights, color="red")
plt.ylabel('No of Departure Delayed Flights')
plt.xlabel('Carriers')
plt.title('No. Of Departure Delays per Carrier')
plt.xticks(ind, flights_df_filtered_grouped.carrier)
plt.show()
```

```python
#Fitering dataset for flights having arr_delay>0
flights_df_filtered = flights_df[(flights_df.arr_delay>0) & np.isfinite(flights_df['arr_time'])]  

#Grouping by carrier and getting the count
flights_df_filtered_grouped = flights_df_filtered.groupby([flights_df_filtered.carrier]).size().reset_index(name='ArrDelayedFLights')
print (flights_df_filtered_grouped)

#Plotting Arrival Delayed Flights for each Carrier
ind = np.arange(len(flights_df_filtered_grouped.carrier))
plt.bar(ind, flights_df_filtered_grouped.ArrDelayedFLights, color="red")
plt.ylabel('No of Arrival Delayed Flights')
plt.xlabel('Carriers')
plt.title('No. Of Arrival Delays per Carrier')
plt.xticks(ind, flights_df_filtered_grouped.carrier)
plt.show()
```

```python
#Fitering dataset for flights having arr_delay>0
flights_df_filtered = flights_df[(flights_df.arr_delay>0) & np.isfinite(flights_df['arr_time'])]  

#Grouping by carrier and getting the mean arrival delay
df_mean_arr_delay = flights_df_filtered.groupby([flights_df_filtered.carrier])['arr_delay'].mean().reset_index(name='Mean_Arrival_Delay')
print (df_mean_arr_delay)

#Plotting Average Arrival Delay of Flights for each Carrier
ind = np.arange(len(df_mean_arr_delay.carrier))
plt.bar(ind, df_mean_arr_delay.Mean_Arrival_Delay, color="blue")
plt.ylabel('Mean_Arrival_Delay')
plt.xlabel('Carriers')
plt.title('Average Arrival Delay per Carrier')
plt.xticks(ind, df_mean_arr_delay.carrier)
plt.show()
```

```python
#Fitering dataset for flights having dep_delay>0
flights_df_filtered = flights_df[(flights_df.dep_delay>0) & np.isfinite(flights_df['dep_time'])]  

#Grouping by carrier and getting the mean departure delay
df_mean_dep_delay = flights_df_filtered.groupby([flights_df_filtered.carrier])['dep_delay'].mean().reset_index(name='Mean_Departure_Delay')
print (df_mean_dep_delay)

#Plotting Average Departure Delay of Flights for each Carrier
ind = np.arange(len(df_mean_dep_delay.carrier))
plt.bar(ind, df_mean_dep_delay.Mean_Departure_Delay, color="blue")
plt.ylabel('Mean_Departure_Delay')
plt.xlabel('Carriers')
plt.title('Average Departure Delay per Carrier')
plt.xticks(ind, df_mean_dep_delay.carrier)
plt.show()
```

```python
#grouping flights df to calculate mean departure delay over the months
grouped_data = pd.DataFrame(flights_df.groupby([flights_df.month])['dep_delay'].mean().reset_index(name='Mean_Delay_Departure'))
print(grouped_data)

#Plotting Mean Departure delay over the months
plt.plot(grouped_data.month, grouped_data.Mean_Delay_Departure,'')
plt.xlim(1, 12)
plt.ylabel('Mean Delay Departure Time (minutes)')
plt.xlabel('Months')
plt.title('Mean Departure delay over the months')
plt.show()
```

```python
#Filtering flights to capture only delayed flights 
filtered_flights_df = flights_df[flights_df.dep_delay > 0]

#Grouping by month to find the count of delays
grouped_data = pd.DataFrame(filtered_flights_df.groupby([filtered_flights_df.month])['dep_delay'].size().reset_index(name='Count'))
print(grouped_data)

#Plotting Number of delays per Month
plt.bar(grouped_data.month, grouped_data.Count, color="red")
plt.ylabel('Number of Delays')
plt.xlabel('Months')
plt.title('Number of delays per Month')
plt.show()
```

```python
#Aircraft Performance: distance-wise analysis
df_aircraft_distance = flights_df.groupby([flights_df.tailnum])['distance'].sum().reset_index(name='Total_Distance_Travelled')
#print (df_aircraft_distance)
print (df_aircraft_distance.sort_values('Total_Distance_Travelled').head(10))
print (df_aircraft_distance.sort_values('Total_Distance_Travelled').tail(10))
```

```python
#Fitering dataset for flights meeting cancellation crietria
flights_df_filtered_carrier = flights_df[flights_df.dep_time.isnull() & flights_df.arr_time.isnull() & flights_df.dep_delay.isnull() & flights_df.arr_delay.isnull() & flights_df.air_time.isnull()]

#Grouping by carrier and getting the count of cancelled flights
flights_df_filtered_grouped = flights_df_filtered_carrier.groupby([flights_df_filtered_carrier.carrier]).size().reset_index(name='Cancelled_Flights')
flights_df_filtered_grouped

#Grouping by carrier to find the total flight count for each carrier
flights_df_filtered_grouped_total = flights_df.groupby([flights_df.carrier]).size().reset_index(name='Total_Flights')
flights_df_filtered_grouped_total

#Merging both the dataframes
flight_df_merged = pd.merge(flights_df_filtered_grouped, flights_df_filtered_grouped_total, on='carrier', how='right')

#Calculating Percentage_Cancellation 
flight_df_merged['Percentage_Cancellation'] = (flight_df_merged.Cancelled_Flights/flight_df_merged.Total_Flights)*100

#Calculating Average_Cancellation 
flight_df_merged['Average_Cancellation'] = (flight_df_merged.Cancelled_Flights/12)

flight_df_merged = flight_df_merged.fillna(0)
print(flight_df_merged)

#Plotting Percentage_Cancellation per Carrier through 2013
ind = np.arange(len(flight_df_merged.carrier))
plt.bar(ind, flight_df_merged.Percentage_Cancellation, color="yellow")
plt.ylabel('Percentage_Cancellation')
plt.xlabel('Carriers')
plt.title('Percentage_Cancellation per Carrier')
plt.xticks(ind, flight_df_merged.carrier)
plt.show()

#Plotting Average_Cancellation per Carrier through 2013
ind = np.arange(len(flight_df_merged.carrier))
plt.bar(ind, flight_df_merged.Average_Cancellation, color="black")
plt.ylabel('Average_Cancellation')
plt.xlabel('Carriers')
plt.title('Average_Cancellation per Carrier over the year')
plt.xticks(ind, flight_df_merged.carrier)
plt.show()
```
